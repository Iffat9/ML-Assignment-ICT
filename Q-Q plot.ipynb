import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Q-Q plot
## 1-symmetric distribution

```python
x=np.random.normal(0,1,200)
xo=np.sort(x)
y=np.linspace(1,200,200)
n=len(x)
p=((y-0.5)/n)
q=stats.norm.ppf(p)
```

##  using continuity correction method

```python
plt.plot(p,xo,'ro',label='data_quantile')
plt.plot(p,q,'go',label='theoretical_quantile')
plt.xlabel('theoretical quantiles')
plt.ylabel('ordered values')
plt.legend()
```

## using inbuilt function

```python
stats.probplot(xo,plot=plt)
```

## 2-positively skewed distribution 

```python
x=np.random.exponential(1,200)
xo=np.sort(x)
p=((y-0.5)/n)
```

```python
q=stats.expon.ppf(p)
```

## using continuity correction

```python
plt.plot(p,xo,'ro',label='data_quantile')
plt.plot(p,q,'bo',label='theoretical_quantile')
plt.xlabel('theoretical quantiles')
plt.ylabel('ordered values')
plt.legend()
```

## using inbuilt function

```python
stats.probplot(xo,dist='expon',plot=plt)
```

## negatively skewed distribution

```python
x=np.random.beta(2,1,200)
xo=np.sort(x)
p=((y-0.5)/n)
q=stats.beta.ppf(p,2,1)
```

```python
plt.plot(p,xo,'ro',label='data_quantile')
plt.plot(p,q,'bo',label='theoretical_quantile')
plt.xlabel('theoretical quantiles')
plt.ylabel('ordered values')
plt.legend()
```

```python
stats.probplot(xo,(2,1),dist='beta',plot=plt)
```

## 4-Heavily tailed distribution 

```python
x=np.random.standard_cauchy(200)
xo=np.sort(x)
p=((y-0.5)/n)
q=stats.cauchy.ppf(p)
```

```python
plt.plot(p,xo,'bo',label='data_quantile')
plt.plot(p,q,'go',label='theoretical_quantile')
plt.xlabel('theoretical quantiles')
plt.ylabel('ordered values')
plt.legend()
```

```python
stats.probplot(xo,dist='cauchy',plot=plt)
```
