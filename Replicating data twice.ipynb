### 1-create a case study in which you generate response of size n and fit a regression line between y and x.
### 2-create another dataset by replecating the same dataset twice .
### again fit a data line estimate the coefficients ,compare the confidence interval for the coefficients in both the cases. 
### do this several times.can you find any pattern?

```python
import numpy as np

import math

from scipy import stats

import pandas as pd

import matplotlib.pyplot as plt

```

```python
b_0=0.5
b_1=1.4
sigma=1
x=np.linspace(1,10,15)
n=len(x)
rep=500
est_b_0=np.zeros(rep)
est_b_1=np.zeros(rep)

f = plt.figure(figsize=(10,5))
f.add_subplot(2,2,1)
np.random.seed(123)
for i in range(rep):
    y=b_0+b_1*x+np.random.normal(0,1,n)
    plt.plot(x,y,'o')
    lm=stats.linregress(x,y)
    line=lm.slope*x+lm.intercept
    plt.plot(x,y,'go',x,line)
    plt.xlabel('x')
    plt.ylabel('y')
    est_b_0[i]=lm.intercept
    est_b_1[i]=lm.slope
vr_b0=sigma**2*((1/n)+((np.mean(x))**2)/sum((x-np.mean(x))**2))
vr_b0
q=np.sqrt(vr_b0)
var_b1=sigma**2/sum((x-np.mean(x))**2)
q1=np.sqrt(var_b1)
import seaborn as sns
f.add_subplot(2,2,2)
sns.distplot(est_b_0,bins=10)
plt.plot(np.mean(est_b_0),0, marker="*",markersize=10, markerfacecolor='r')
plt.plot(b_0,0,marker="*",markersize=10,markerfacecolor='g')
plt.vlines(est_b_0[i]-1.96*q, ymin=0,ymax=1,color="green",label='conf int')
plt.vlines(est_b_0[i]+1.96*q, ymin=0,ymax=1,color="green" )
plt.legend()
plt.title(r"$\beta_0$")
plt.xlabel('est_b_0')
plt.ylabel('Density')
f.add_subplot(2,2,3)
sns.distplot(est_b_1,bins=10)
plt.plot(np.mean(est_b_1),0, marker="*",markersize=10, markerfacecolor='r')
plt.plot(b_1,0,marker="*",markersize=10,markerfacecolor='g')
plt.vlines(est_b_1[i]-1.96*q1, ymin=0,ymax=4,color="green",label='conf int' )
plt.vlines(est_b_1[i]+1.96*q1, ymin=0,ymax=4,color="green" )
plt.legend()
plt.xlabel('est_b_1')
plt.ylabel('Density')
plt.title(r"$\beta_1$")
plt.tight_layout()
```

```python
b_0=0.5
b_1=1.4
sigma=1
x=np.linspace(1,10,15)
rep=500
est_b_0=np.zeros(rep)
est_b_1=np.zeros(rep)

f = plt.figure(figsize=(10,5))
f.add_subplot(2,2,1)
np.random.seed(123)
for i in range(rep):
    x1=x
    x2=np.hstack((x,x1))
    n=len(x2)
    y=b_0+b_1*x1+np.random.normal(0,1,15)
    y1=y
    y2=np.hstack((y,y1))
    plt.plot(x2,y2,'o')
    lm=stats.linregress(x,y)
    line=lm.slope*x+lm.intercept
    plt.plot(x2,y2,'go',x,line)
    plt.xlabel('x2')
    plt.ylabel('y2')
    est_b_0[i]=lm.intercept
    est_b_1[i]=lm.slope
vr_b0=sigma**2*((1/n)+((np.mean(x2))**2)/sum((x2-np.mean(x2))**2))
vr_b0
q=np.sqrt(vr_b0)
var_b1=sigma**2/sum((x2-np.mean(x2))**2)
q1=np.sqrt(var_b1)
import seaborn as sns
f.add_subplot(2,2,2)

sns.distplot(est_b_0,bins=10)
plt.plot(np.mean(est_b_0),0, marker="*",markersize=10, markerfacecolor='r')
plt.plot(b_0,0,marker="*",markersize=10,markerfacecolor='g')
plt.vlines(est_b_0[i]-1.96*q, ymin=0,ymax=1,color="green",label='conf int')
plt.vlines(est_b_0[i]+1.96*q, ymin=0,ymax=1,color="green" )
plt.legend()
plt.title(r"$\beta_0$")
plt.xlabel('est_b_0')
plt.ylabel('Density')
f.add_subplot(2,2,3)
sns.distplot(est_b_1,bins=10)
plt.plot(np.mean(est_b_1),0, marker="*",markersize=10, markerfacecolor='r')
plt.plot(b_1,0,marker="*",markersize=10,markerfacecolor='g')
plt.vlines(est_b_1[i]-1.96*q1, ymin=0,ymax=4,color="green",label='conf int' )
plt.vlines(est_b_1[i]+1.96*q1, ymin=0,ymax=4,color="green" )
plt.legend()
plt.xlabel('est_b_1')
plt.ylabel('Density')
plt.title(r"$\beta_1$")
plt.tight_layout()
```

### by looking at both the plots of different dataset it is visible that the confidence interval of the coefficients of  dataset which is obtained by replicating the first one twice is reduced.
